## Visualization Basics

### Comparison plots

#### Bar Plot

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
languages = ['C', 'C++', 'Java', 'Python', 'R']
students = [23, 17, 29, 35, 12]
```

```python
plt.bar(languages, students)
plt.title('Student Distribution by Language Popularity')
plt.xlabel('Programming Languages')
plt.ylabel('Students Enrolled')
```

```python
plt.barh(languages, students)
plt.title('Student Distribution by Language Popularity')
plt.ylabel('Programming Languages')
plt.xlabel('Students Enrolled')
```

```python
languages = ['C', 'C++', 'Java', 'Python', 'R']
students_2022 = [23, 17, 29, 35, 12]
students_2023 = [20, 15, 24, 40, 15]


plt.bar(languages, students_2022, alpha=0.5)
plt.bar(languages, students_2023, alpha=0.5)
plt.title('Student Distribution by Language Popularity')
plt.xlabel('Programming Languages')
plt.ylabel('Students Enrolled')
plt.legend(['Student 2022', 'Student 2023'])
```

```python
index = np.arange(len(languages))
width = 0.35

plt.bar(index, students_2022, width=width, label='2022')
plt.bar(index+width, students_2023, width=width, label='2023')
plt.title('Student Distribution by Language Popularity')
plt.xlabel('Programming Languages')
plt.ylabel('Students Enrolled')
plt.legend()

plt.xticks((index+width/2), languages)
```

```python
classes = ['DV-1', 'DV-2', 'DL-1', 'DL-2', 'CA']
students__mean_score = [23, 17, 29, 35, 12]
students_variance = [1, 2, 5, 7, 3]
```

```python
plt.bar(classes, students__mean_score, yerr=students_variance)

plt.title('Student Distribution')
plt.xlabel('Courses')
plt.ylabel('Students Average')
```

```python
y1 = [2, 4, 5, 8]
y2 = [1, 5, 7, 5]

plt.plot(y1, marker='D', markersize=8, markerfacecolor='orange', color='green', markeredgecolor='blue',
            linestyle='-.', linewidth=2, label='y1')

plt.plot(y2, marker='*', markersize=8, markerfacecolor='green', color='blue', markeredgecolor='red',
            linestyle='--', linewidth=2, label='y2')

plt.legend()
```

```python
languages = ['C', 'C++', 'Java', 'Python', 'R']
students_2022 = [23, 17, 29, 35, 12]
students_2023 = [20, 15, 24, 40, 15]
```

```python
fig, ax = plt.subplots(1, 2) # 1-row 1-col
ax[1].plot(y1, marker='D', markersize=8, markerfacecolor='orange', color='green', markeredgecolor='red',
            linestyle='-.', linewidth=2, label='Y1')

ax[0].plot(y2, marker='*', markersize=8, markerfacecolor='green', color='blue', markeredgecolor='red',
            linestyle='--', linewidth=2, label='y2')
```

```python
fig, ax = plt.subplots(1, 1) # 1-row 1-col
ax.plot(students_2022, marker='D', markersize=8, markerfacecolor='orange', color='green', markeredgecolor='red',
            linestyle='-.', linewidth=2, label='Y1')

ax.plot(students_2023, marker='*', markersize=8, markerfacecolor='green', color='blue', markeredgecolor='red',
            linestyle='--', linewidth=2, label='y2')

ax.set_xticks(range(len(languages)))
ax.set_xticklabels(languages, rotation=60)
plt.legend()
```

```python
fig, bx = plt.subplots(1, 2, figsize=(18, 6))
# bar plots
index = np.arange(len(languages))
width = 0.35
bx[0].bar(index, students_2022, width=width, label='2022')
bx[0].bar(index+width, students_2023, width=width, label='2023')
bx[0].set_xticks((index+width/2), languages)

bx[0].set_title('Bar Graph', fontsize=16)
bx[0].set_xlabel('Programming Language')
bx[0].set_ylabel('Students Enrolled')
bx[0].legend()

# line plots
bx[1].plot(students_2022, marker='D', markersize=8, markerfacecolor='orange', color='green', markeredgecolor='red',
            linestyle='-.', linewidth=2, label='2022')

bx[1].plot(students_2023, marker='*', markersize=8, markerfacecolor='green', color='blue', markeredgecolor='red',
            linestyle='--', linewidth=2, label='2023')

bx[1].set_xticks((index+width/2), languages)
bx[1].set_title('Line Chart', fontsize=16, loc='left')
bx[1].set_xlabel('Programming Language')
bx[1].set_ylabel('Students Enrolled')
bx[1].legend()

fig.suptitle('Student\'s Language Popularity by Year', fontsize=24, x=0.5, y=1.05)
```

```python
fig, bx = plt.subplots(1, 1, figsize=(10, 6))
# bar plots
index = np.arange(len(languages))
width = 0.35
bx.bar(index, students_2022, width=width, label='2022')
bx.bar(index+width, students_2023, width=width, label='2023')

# line plots
bx.plot(index, students_2022, marker='D', markersize=8, markerfacecolor='orange', color='green', markeredgecolor='red',
            linestyle='-.', linewidth=2, label='2022')

bx.plot(index+width, students_2023, marker='*', markersize=8, markerfacecolor='green', color='blue', markeredgecolor='red',
            linestyle='--', linewidth=2, label='2023')

bx.set_xticks((index+width/2), languages)
bx.set_title('Bar Graph', fontsize=16)
bx.set_xlabel('Programming Language')
bx.set_ylabel('Students Enrolled')
bx.legend()


fig.suptitle('Student\'s Language Popularity by Year', fontsize=24, x=0.5, y=1.05)
```
