### Visualization with Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.isnull().sum()
```

### Matplotlib

```python
plt.scatter(df.sepal_length, df.sepal_width)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

### Seaborn

```python
sns.relplot(x='sepal_length', y='sepal_width', data=df)
```

```python
sns.relplot(x='sepal_length', y='sepal_width', data=df, hue='species')
```

```python
sns.relplot(x='sepal_length', y='sepal_width', data=df, hue='species', kind='line')
```

```python
sns.relplot(x='sepal_length', y='sepal_width', data=df, hue='species', kind='scatter')
sns.relplot(x='sepal_length', y='sepal_width', data=df, hue='species', kind='line')
```

```python
# Removing confidence interval from the graph
sns.relplot(x='sepal_length', y='sepal_width', data=df, hue='species', kind='line', ci=False)
```

```python
sns.relplot(x='sepal_length', y='sepal_width', data=df, hue='species', kind='line', errorbar=('ci', False))
```

### Regression

```python
# Linear Regression
sns.lmplot(x='sepal_length', y='sepal_width', data=df)
```

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=df, hue='species')
```

```python
sns.lmplot(x='petal_length', y='petal_width', data=df)
```

```python
sns.lmplot(x='petal_length', y='petal_width', data=df, hue='species')
```

```python
sns.lmplot(x='petal_length', y='petal_width', data=df, row='species')
```

```python
sns.lmplot(x='petal_length', y='petal_width', data=df, col='species')
```

### regression plot

```python
plt.rcParams['figure.figsize'] = (12, 9)
```

```python
# First order equation y=mx+c
sns.regplot(x='sepal_length', y='sepal_width', data=df)
```

```python
# second order equation y=ax2+bx+c
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=2)
```

```python
# third order equation y=ax3+bx2+cx+d
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=3)
```

```python
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=5)
```

```python
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=7)
```

```python
plt.figure(figsize=(12, 9))
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=1, label='Order-1')
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=2, label='Order-2')
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=3, label='Order-3')
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=4, label='Order-4')
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=5, label='Order-5')
sns.regplot(x='sepal_length', y='sepal_width', data=df, order=7, label='Order-7')
plt.legend()
```

```python
plt.figure(figsize=(12, 9))
sns.regplot(x='petal_length', y='petal_width', data=df, order=1, label='Order-1')
sns.regplot(x='petal_length', y='petal_width', data=df, order=2, label='Order-2')
sns.regplot(x='petal_length', y='petal_width', data=df, order=3, label='Order-3')
sns.regplot(x='petal_length', y='petal_width', data=df, order=4, label='Order-4')
sns.regplot(x='petal_length', y='petal_width', data=df, order=5, label='Order-5')
sns.regplot(x='petal_length', y='petal_width', data=df, order=7, label='Order-7')
plt.legend()
```
