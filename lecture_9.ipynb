## Plotly

```python
import plotly.express as px
```

```python
fig = px.bar(x=['Python', 'R', 'Java', 'C', 'C++'], 
             y=[45, 15, 25, 7, 10])
fig.show()
```

```python
fig = px.scatter(x=range(3), y=range(3))
fig.show()
```

#### Working with Dataset

```python
gapminder = px.data.gapminder()
gapminder.head(30)
```

```python
gapminder.shape
```

```python
gapminder['year']
```

```python
gm_2007 = gapminder[gapminder['year'] == 2007]
gm_2007.head()
```

```python
gm_1952 = gapminder.query("year == 1952")
gm_1952.head()
```

```python

gapminder.query('year > 1952 and year < 1962')
```

```python
gapminder[(gapminder['year'] > 1952) & (gapminder['year'] < 1962)]
```

```python
fig = px.scatter(gm_1952, x = 'gdpPercap', y='lifeExp')
fig.show()
```

```python
fig = px.scatter(gm_1952, x = 'gdpPercap', y='lifeExp',
                log_x=True)
fig.show()
```

```python
fig = px.scatter(gm_1952, x = 'gdpPercap', y='lifeExp',
                log_x=True, hover_name='country')
fig.show()
```

```python
fig = px.scatter(gm_1952, x = 'gdpPercap', y='lifeExp',
                log_x=True, hover_name='country', color='continent')
fig.show()
```

```python
fig = px.scatter(gm_1952, x = 'gdpPercap', y='lifeExp',
                log_x=True, hover_name='country', color='continent',
                size='pop')
fig.show()
```

```python
fig_1952 = px.scatter(gm_1952, x = 'gdpPercap', y='lifeExp',
                log_x=True, hover_name='country', color='continent',
                size='pop', size_max=60, 
                title='A Plotly Express Gapminder Figure for 1952: LifeExp vs GDP-per-capita')
fig_1952.show()
```

```python
fig_2007 = px.scatter(gm_2007, x = 'gdpPercap', y='lifeExp',
                log_x=True, hover_name='country', color='continent',
                size='pop', size_max=60, 
                title='A Plotly Express Gapminder Figure for 2007: LifeExp vs GDP-per-capita')
fig_2007.show()
```
